# Matplotlib

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

[matplotlib website](https://matplotlib.org/)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 10])
y = np.array([0, 100])
plt.plot(x,y)
```

#### Basic Matplotlib Commands

```python
x = np.linspace(0, 5, 20)
y = x ** 2
```

```python
plt.plot(x, y, color='green', marker='*',  markersize=10)
```

```python
plt.plot(x,y,'r+')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('plot x**2')
plt.show()
```

#### Creating Multiplots

```python
plt.plot(x,y,'r')
plt.plot(y,x,'b')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r*')
plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

## Matplotlib Object Oriented Method

OOP gives more control and customization of your plots.

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.5,0.5,0.5])
axes.plot(x,y,'g')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
type(fig)
```

```python
fig = plt.figure(facecolor='pink')

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y,'g')
axes1.set_xlabel('X_Label_axes1')
axes1.set_ylabel('Y_Label_axes1')
axes1.set_title('Title1')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(x,y,'y')
axes2.set_xlabel('X_Label_axes2')
axes2.set_ylabel('Y_Label_axes2')
axes2.set_title('Title2')
```

```python
fig = plt.figure(facecolor='orange')
fig2 = plt.figure(facecolor='purple')

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y,'g')
axes1.set_xlabel('X_Label_axes1')
axes1.set_ylabel('Y_Label_axes1')
axes1.set_title('Title1')

axes2 = fig2.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(x,y,'y')
axes2.set_xlabel('X_Label_axes2')
axes2.set_ylabel('Y_Label_axes2')
axes2.set_title('Title2')
```

### subplots()

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.grid(True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(y,x)
axes[1].set_title('Plot 2')
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y, 'b')
```

```python
fig, axes = plt.subplots(figsize=(12,4))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

### Combine

```python
fig, axes = plt.subplots(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(y,x,'b--')
plt.subplot(2,2,3 )
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(y,x,'p--')
fig.savefig('myfile2.png')
```

### Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x**2')
ax.plot(x,x**3, label='x**3')
ax.legend(loc=0) #let matplotlib decide the optimal location 
```

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
```

### Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots()
ax.plot(x,x+1,color='blue',alpha=0.5) # half-transparant
```

##### Line and marker styles

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```
